The plot here shows extremely high count for non-fraud transactions and very low count for fraud transactions. Now we will check how much percentage of total transactions are non-fraud

We to built the model on a balanced dataset which is later used for entire dataset to make predictions

```python
# Since most of our data has already been scaled we should scale the columns that are left to scale (Amount and Time)
from sklearn.preprocessing import RobustScaler

# RobustScaler is less prone to outliers.
rob_scaler = RobustScaler()

df['scaled_amount'] = rob_scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df['scaled_time'] = rob_scaler.fit_transform(df['Time'].values.reshape(-1,1))

df.drop(['Time','Amount'], axis=1, inplace=True)
cols = df.columns.tolist()
cols = cols[-2:] + cols[:-2]
df = df[cols]
df.head()
```

#### Balancing Data

#### Random under sampling   
Here we have more non-fraud than fraud. So in order to make balanced dataset we reduce non-fraud transaction so that those will be equal to no. of fraud transaction. But one main issue is our classification models will not perform as accurate as we would like to since there is a great loss of information (bringing 492 non-fraud transaction from 284,315 non-fraud transaction)

```python
# Since our classes are highly skewed we should make them equivalent in order to have a normal distribution of the classes.

# Lets shuffle the data before creating the subsamples

df = df.sample(frac=1)

# amount of fraud classes 492 rows.
fraud_df = df.loc[df['Class'] == 1]
non_fraud_df = df.loc[df['Class'] == 0][:492]

normal_distributed_df = pd.concat([fraud_df, non_fraud_df])

# Shuffle dataframe rows
new_df = normal_distributed_df.sample(frac=1, random_state=42)

new_df.head()
```

```python
new_df.hist(column='Class')
plt.show()
```

### Correlation Matrix  
It is important to check if there are any features which influences heavily to tell if a transaction is fraud or not. It is important to use correct dataframe (here subsampled dataframe which is new_df) to check if we have positive or negative correlation with the dependent variable. 

```python
# Make sure we use the subsample in our correlation

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(24,20))
#fig, ax = plt.subplots()

# Entire DataFrame
corr = df.corr()
sns.heatmap(corr, fmt="d", annot_kws={'size':20}, ax=ax1)
ax1.set_title("Imbalanced Correlation Matrix \n (don't use for reference)", fontsize=14)


sub_sample_corr = new_df.corr()
sns.heatmap(sub_sample_corr, fmt="d", annot_kws={'size':20}, ax=ax2)
ax2.set_title('SubSample Correlation Matrix \n (use for reference)', fontsize=14)
plt.show()
```

With above heatmap we can say V2,V4,V11,V9 are highly correlated with Class variable. Similary, V3,V9,V10,V12,V14,V16,V17 are least correlated with Class variable.
